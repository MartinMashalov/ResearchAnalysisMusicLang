```python
#correlation function here
def correlations(data, column1, column2): 
    x = np.array(list(data[column1]))
    y = np.array(list(data[column2]))
    result = scipy.stats.pearsonr(x, y)
    return result

```

```python
lang = pd.read_csv("CleanDataLangMusic.csv")

def fix_practice():
    for idx, row in lang_filtered.iterrows(): 
        if lang.loc[idx, 'Length of Practice'] == "None": 
            lang.loc[idx, 'Length of Practice'] = 0
        elif lang.loc[idx, 'Length of Practice'] == "1 year or so":
            lang.loc[idx, 'Length of Practice'] = 1
        elif lang.loc[idx, 'Length of Practice'] == "2 years":
            lang.loc[idx, 'Length of Practice'] = 2
        elif lang.loc[idx, 'Length of Practice'] == "Around 3 years.":
            lang.loc[idx, 'Length of Practice'] = 3
        elif lang.loc[idx, 'Length of Practice'] == "Guitar 4 years, Piano 7 years, Vocal 7 years":
            lang.loc[idx, 'Length of Practice'] = 4
        elif lang.loc[idx, 'Length of Practice'] == "5 years":
            lang.loc[idx, 'Length of Practice'] = 5
        elif lang.loc[idx, 'Length of Practice'] == "6 years":
            lang.loc[idx, 'Length of Practice'] = 6
        elif lang.loc[idx, 'Length of Practice'] == "7 years":
            lang.loc[idx, 'Length of Practice'] = 7
        elif lang.loc[idx, 'Length of Practice'] == "9 years":
            lang.loc[idx, 'Length of Practice'] = 9
        elif lang.loc[idx, 'Length of Practice'] == "12 years":
            lang.loc[idx, 'Length of Practice'] = 12
        elif lang.loc[idx, 'Length of Practice'] == "13 years":
            lang.loc[idx, 'Length of Practice'] = 13

fix_practice()

lang['Rounded_Musical_IQ'] = lang['MusicalIQ'].apply(roundup_ten)
lang['Rounded_Beat_Score'] = lang['Beat Alignment Score'].apply(roundup_ten)
lang['Rounded_Melodic'] = lang['Melodic Descrimination Score'].apply(roundup_ten)
lang['Rounded_Perception_Score'] = lang['Mistuning Perception Score'].apply(roundup_ten)

lang['Rounded_Musical_IQ_20'] = lang['MusicalIQ'].apply(roundup_20)
lang['Rounded_Beat_Score_20'] = lang['Beat Alignment Score'].apply(roundup_20)
lang['Rounded_Melodic_20'] = lang['Melodic Descrimination Score'].apply(roundup_20)
lang['Rounded_Perception_Score_20'] = lang['Mistuning Perception Score'].apply(roundup_20)

column_list1 = ['Rounded_Musical_IQ', 'Rounded_Beat_Score', 'Rounded_Melodic', 'Rounded_Perception_Score']
column_list = ['Rounded_Musical_IQ_20', 'Rounded_Beat_Score_20', 'Rounded_Melodic_20', 'Rounded_Perception_Score_20']
comparison_array = ['English First Language', 'Language Home Yes/No', 'Languages at Home Number', 'Perfect Pitch', 'Instrument', 'Length of Practice']

def create_correlations():
    for comparison in comparison_array: 
        print(' ')
        print("---------------{}-------------".format(comparison))
        try: 
            for item in column_list1: 
                print(correlations(lang, item, comparison), item)
        except: 
            continue
            
#(r value, p value)
create_correlations()
```

```python
#ANOVA test
import statsmodels.api as sm 
from statsmodels.formula.api import ols 

#print(lang.head())
lang.rename(columns={"Length of Practice": "Practice"}, inplace=True)


#Rounded_Beat_Score + Rounded_Melodic + Rounded_Perception_Score
comparison_array = ['Rounded_Musical_IQ', 'Rounded_Beat_Score', 'Rounded_Perception_Score', 'Rounded_Melodic']

for comp in comparison_array: 
    print("-----------------{}-----------------".format(comp))
    data_string = "{} ~ Age + Grade + Practice".format(comp)
    mod = ols(data_string, data=lang).fit()
    anova_table = sm.stats.anova_lm(mod, typ=3)
    print(anova_table)
    print("\n")
```
